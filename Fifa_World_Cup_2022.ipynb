```python
team_possession = data.groupby('team1')['possession team1'].mean().append(data.groupby('team2')['possession team2'].mean())
team_colors = ['red', 'green', 'blue', 'orange', 'purple','yellow','pink','grey'] 
plt.figure(figsize=(12, 6))
team_possession.plot(kind='bar',color= team_colors)
plt.xlabel('Team')
plt.ylabel('Average Possession')
plt.title('Average Possession by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a list of all unique team names
all_teams = pd.unique(data[['team1', 'team2']].values.ravel())

# Calculate number of goals for each team
team_goals = pd.Series(0, index=all_teams)
team_goals += data.groupby('team1')['number of goals team1'].sum()
team_goals += data.groupby('team2')['number of goals team2'].sum()
team_colors = ['red', 'green', 'blue', 'orange', 'purple','yellow','pink','grey'] 
plt.figure(figsize=(12, 6))
team_goals.plot(kind='bar',color=team_colors)
plt.xlabel('Team')
plt.ylabel('Number of Goals')
plt.title('Number of Goals by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a list of all unique team names
all_teams = pd.unique(data[['team1', 'team2']].values.ravel())

# Calculate total attempts on goals for each team
team_attempts = pd.Series(0, index=all_teams)
team_attempts += data.groupby('team1')['total attempts team1'].sum()
team_attempts += data.groupby('team2')['total attempts team2'].sum()
team_colors = ['red', 'green', 'blue', 'orange', 'purple','yellow','pink','grey'] 
plt.figure(figsize=(12, 6))
team_attempts.plot(kind='bar',color= team_colors)
plt.xlabel('Team')
plt.ylabel('Total Attempts on Goals')
plt.title('Total Attempts on Goals by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a list of all unique team names
all_teams = pd.unique(data[['team1', 'team2']].values.ravel())

# Calculate total goals inside penalty area by each team
total_goals_inside_penalty_area_team1 = data.groupby('team1')['goal inside the penalty area team1'].sum()
total_goals_inside_penalty_area_team2 = data.groupby('team2')['goal inside the penalty area team2'].sum()

# Calculate total goals outside penalty area by each team
total_goals_outside_penalty_area_team1 = data.groupby('team1')['goal outside the penalty area team1'].sum()
total_goals_outside_penalty_area_team2 = data.groupby('team2')['goal outside the penalty area team2'].sum()

# Combine the goals inside and outside the penalty area for each team
total_goals_inside_penalty_area = total_goals_inside_penalty_area_team1.add(total_goals_inside_penalty_area_team2, fill_value=0)
total_goals_outside_penalty_area = total_goals_outside_penalty_area_team1.add(total_goals_outside_penalty_area_team2, fill_value=0)

# Create a bar plot for total goals inside and outside the penalty area
fig, ax = plt.subplots(figsize=(10, 6))
total_goals_inside_penalty_area.plot(kind='bar', color='green', alpha=0.7, label='Goals Inside Penalty Area')
total_goals_outside_penalty_area.plot(kind='bar', color='red', alpha=0.7, label='Goals Outside Penalty Area')
ax.set_xlabel('Team')
ax.set_ylabel('Number of Goals')
ax.set_title('Total Goals Inside and Outside the Penalty Area by Team')
ax.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate total assists provided by each team
total_assists_team1 = data.groupby('team1')['assists team1'].sum()
total_assists_team2 = data.groupby('team2')['assists team2'].sum()

# Combine the total assists from both teams
total_assists = total_assists_team1.add(total_assists_team2, fill_value=0)

# Sort the total assists in descending order
total_assists = total_assists.sort_values(ascending=False)

# Create a bar plot for total assists by team
plt.figure(figsize=(10, 6))
total_assists.plot(kind='bar', color='orange')
plt.xlabel('Team')
plt.ylabel('Total Assists')
plt.title('Total Assists by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate switches of play completed by team
team_switches_completed = data.groupby('team1')['switches of play completed team1'].sum()

# Create a bar plot for switches of play completed
plt.figure(figsize=(10, 5))
plt.bar(team_switches_completed.index, team_switches_completed)
plt.xlabel('Team')
plt.ylabel('Switches of Play Completed')
plt.title('Switches of Play Completed by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate set-piece statistics by team
team_set_piece_stats = data.groupby('team1')[['corners team1', 'free kicks team1', 'penalties scored team1', 'goal preventions team1']].sum()

# Create a stacked bar plot for set-piece statistics
team_set_piece_stats.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Set-Piece Statistics by Team')
plt.xticks(rotation=90)
plt.legend(title='Set-Piece Type')
plt.show()
```

```python
# Calculate on-target and off-target attempts for each team
team_on_target_attempts = data.groupby('team1')['on target attempts team1'].sum()
team_off_target_attempts = data.groupby('team1')['off target attempts team1'].sum()

# Create a bar plot for on-target and off-target attempts
plt.figure(figsize=(10, 5))
plt.bar(team_on_target_attempts.index, team_on_target_attempts, label='On-Target Attempts')
plt.bar(team_off_target_attempts.index, team_off_target_attempts, label='Off-Target Attempts')
plt.xlabel('Team')
plt.ylabel('Number of Attempts')
plt.title('On-Target and Off-Target Attempts by Team')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Calculate attempts made in different areas of the field
team_left_channel_attempts = data.groupby('team1')['left channel team1'].sum()
team_center_channel_attempts = data.groupby('team1')['central channel team1'].sum()
team_right_channel_attempts = data.groupby('team1')['right channel team1'].sum()

# Create a bar plot for attempts in different channels
plt.figure(figsize=(10, 5))
plt.bar(team_left_channel_attempts.index, team_left_channel_attempts, label='Left Channel')
plt.bar(team_center_channel_attempts.index, team_center_channel_attempts, label='Center Channel')
plt.bar(team_right_channel_attempts.index, team_right_channel_attempts, label='Right Channel')
plt.xlabel('Team')
plt.ylabel('Number of Attempts')
plt.title('Attempts in Different Channels by Team')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Calculate offers to receive by teams
team_offers_to_receive = data.groupby('team1')['total offers to receive team1'].sum()

# Create a bar plot for offers to receive
plt.figure(figsize=(10, 5))
plt.bar(team_offers_to_receive.index, team_offers_to_receive)
plt.xlabel('Team')
plt.ylabel('Number of Offers to Receive')
plt.title('Offers to Receive by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
# Calculate line breaks attempted and completed by teams
line_breaks_attempted_team1 = data.groupby('team1')['attempted line breaks team1'].sum()
line_breaks_attempted_team2 = data.groupby('team1')['attempted line breaks team2'].sum()
line_breaks_completed_team1 = data.groupby('team1')['completed line breaksteam1'].sum()
line_breaks_completed_team2 = data.groupby('team1')['completed line breaks team2'].sum()

# Create a bar plot for line breaks attempted and completed
plt.figure(figsize=(10, 5))
plt.bar(line_breaks_attempted_team1.index, line_breaks_attempted_team1, label='Team 1 Line Breaks Attempted')
plt.bar(line_breaks_attempted_team2.index, line_breaks_attempted_team2, label='Team 2 Line Breaks Attempted')
plt.bar(line_breaks_completed_team1.index, line_breaks_completed_team1, label='Team 1 Line Breaks Completed')
plt.bar(line_breaks_completed_team2.index, line_breaks_completed_team2, label='Team 2 Line Breaks Completed')
plt.xlabel('Team')
plt.ylabel('Number of Line Breaks')
plt.title('Line Breaks Attempted and Completed by Team')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Calculate defensive line breaks attempted and completed by teams
defensive_line_breaks_attempted_team1 = data.groupby('team1')['attempted defensive line breaks team1'].sum()
defensive_line_breaks_attempted_team2 = data.groupby('team1')['attempted defensive line breaks team2'].sum()
defensive_line_breaks_completed_team1 = data.groupby('team1')['completed defensive line breaksteam1'].sum()
defensive_line_breaks_completed_team2 = data.groupby('team1')['completed defensive line breaks team2'].sum()

# Create a bar plot for defensive line breaks attempted and completed
plt.figure(figsize=(10, 5))
plt.bar(defensive_line_breaks_attempted_team1.index, defensive_line_breaks_attempted_team1, label='Team 1 Defensive Line Breaks Attempted')
plt.bar(defensive_line_breaks_attempted_team2.index, defensive_line_breaks_attempted_team2, label='Team 2 Defensive Line Breaks Attempted')
plt.bar(defensive_line_breaks_completed_team1.index, defensive_line_breaks_completed_team1, label='Team 1 Defensive Line Breaks Completed')
plt.bar(defensive_line_breaks_completed_team2.index, defensive_line_breaks_completed_team2, label='Team 2 Defensive Line Breaks Completed')
plt.xlabel('Team')
plt.ylabel('Number of Defensive Line Breaks')
plt.title('Defensive Line Breaks Attempted and Completed by Team')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Calculate number of yellow and red cards received by teams
yellow_cards_team1 = data.groupby('team1')['yellow cards team1'].sum()
yellow_cards_team2 = data.groupby('team1')['yellow cards team2'].sum()
red_cards_team1 = data.groupby('team1')['red cards team1'].sum()
red_cards_team2 = data.groupby('team1')['red cards team2'].sum()

plt.figure(figsize=(10, 5))
plt.bar(yellow_cards_team1.index, yellow_cards_team1, label='Team 1 Yellow Cards')
plt.bar(yellow_cards_team2.index, yellow_cards_team2, label='Team 2 Yellow Cards')
plt.bar(red_cards_team1.index, red_cards_team1, label='Team 1 Red Cards')
plt.bar(red_cards_team2.index, red_cards_team2, label='Team 2 Red Cards')
plt.xlabel('Team')
plt.ylabel('Number of Cards')
plt.title('Yellow and Red Cards Received by Team')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
fouls_committed = data.groupby('team1')['fouls against team1'].sum().add(data.groupby('team2')['fouls against team2'].sum(), fill_value=0)
team_colors = ['red', 'green', 'blue', 'orange', 'purple','yellow','pink'] 
plt.figure(figsize=(10, 6))
fouls_committed.plot(kind='bar',color=team_colors)
plt.title('Total Fouls Committed Against Teams')
plt.xlabel('Team')
plt.ylabel('Total Number of Fouls Committed')
plt.show()
```

```python
offsides_violations = data.groupby('team1')['offsides team1'].sum().add(data.groupby('team2')['offsides team2'].sum(), fill_value=0)
team_colors = ['red', 'yellow', 'pink', 'blue','green','orange', 'purple'] 
plt.figure(figsize=(10, 6))
offsides_violations.plot(kind='bar',color= team_colors)
plt.title('Total Offside Violations')
plt.xlabel('Team')
plt.ylabel('Total Number of Offside Violations')
plt.show()
```

```python
crosses_attempted_team1 = data.groupby('team1')['crosses team1'].sum().add(data.groupby('team1')['crosses team2'].sum(), fill_value=0)
crosses_completed_team1 = data.groupby('team1')['crosses completed team1'].sum().add(data.groupby('team1')['crosses completed team2'].sum(), fill_value=0)
team_colors = ['red', 'green', 'blue', 'orange', 'purple']  
plt.figure(figsize=(10, 6))
crosses_attempted_team1.plot(kind='bar', label='Crosses Attempted',color=team_colors)
crosses_completed_team1.plot(kind='bar', label='Crosses Completed',color=team_colors)
plt.title('Crosses Attempted and Completed')
plt.xlabel('Team')
plt.ylabel('Number of Crosses')
plt.legend()
plt.show()
```

```python
passes_made = data.groupby('team1')['passes team1'].sum().add(data.groupby('team2')['passes team2'].sum(), fill_value=0)
team_colors = ['red', 'green', 'blue', 'orange', 'purple'] 
plt.figure(figsize=(10, 6))
passes_made.plot(kind='bar',color=team_colors)
plt.title('Total Number of Passes Made')
plt.xlabel('Team')
plt.ylabel('Total Number of Passes Made')
plt.show()
```

```python
# Calculate average pass completion rate for each team
avg_pass_completion_rate = (data['passes completed team1'] + data['passes completed team2']) / (data['passes team1'] + data['passes team2']) * 100

avg_pass_completion_rate_by_team = avg_pass_completion_rate.groupby(data['team1']).mean()

team_colors = ['red', 'green', 'blue', 'orange', 'purple']  

plt.figure(figsize=(10, 6))
plt.bar(avg_pass_completion_rate_by_team.index, avg_pass_completion_rate_by_team, color=team_colors)
plt.xlabel('Team')
plt.ylabel('Average Pass Completion Rate (%)')
plt.title('Average Pass Completion Rate by Team')
plt.xticks(rotation=90)
plt.show()
```
